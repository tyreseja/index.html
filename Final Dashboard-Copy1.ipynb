```python
#first bar chart

@app.callback(
    Output(component_id='our_graph', component_property='figure'),
    [Input(component_id='my_dropdown', component_property='value')]
)

def build_graph(column_chosen):
    dff=df
    fig = px.bar(dff,x='Location',y=column_chosen,color='Location')
    fig.update_layout(title={'text':"Analysis of Detroit's MLK Jr BLvd",
                      'font':{'size':28},'x':0.5,'xanchor':'center'})
    return fig
```

```python
#second bar chart

@app.callback(
    Output(component_id='our_graph2', component_property='figure'),
    [Input(component_id='my_dropdown2', component_property='value')]
)

def build_graph(column_chosen):
    dff=df
    fig = px.bar(dff,x='Location',y=column_chosen,color='Location')
   
    return fig
```

```python
#third bar chart(airqualty)

@app.callback(
    Output(component_id='our_graph3', component_property='figure'),
    [Input(component_id='my_dropdown3', component_property='value')]
)

def build_graph(column_chosen):
    dff3=df3
    fig = px.bar(dff3,x='Location',y=column_chosen,color='Location')
   
    return fig
```

```python
#fourth chart (population pie chart)

@app.callback(
    Output(component_id='our_graph4', component_property='figure'),
    [Input(component_id='my_dropdown4', component_property='value')]
)
                
def build_graph(column_chosen): 
    dff2=df2
    fig = px.pie(dff2,names='Unnamed: 0',values= column_chosen,) 
    fig.update_traces(textinfo='percent+label')

   
    return fig
```
